- Scipy -> Implementa algoritmos de optimización (https://scipy.org/)

```python
#Ajustando una curva de datos
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.1, 3.5, 7.2, 13.1, 21.3])

# Función a ajustar
def func(x, a, b, c):
    return a * x**2 + b * x + c

# Ajuste
popt, pcov = curve_fit(func, x, y)

#Imprimir Parameters and Covariance
print(f"Parameters {popt}")
print(f"Covariance {pcov}")

# Coeficientes
a, b, c = popt

# Gráfica
plt.plot(x, y, 'ro', label='Datos')
plt.plot(x, func(x, a, b, c), label='Ajuste')
plt.legend()
plt.show()
```
