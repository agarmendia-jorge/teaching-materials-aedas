VisualizaciÃ³n de Datos
- matplotlib -> Es una herramienta para visualizar datos (https://matplotlib.org/)

```python
#Ejemplo

#Line chart
from matplotlib import pyplot as plt

years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
students_eps = [200, 251, 241, 596, 126, 147, 152, 210, 352, 521, 425]
students_law = [190, 260, 256, 59, 542, 142, 180, 456, 56, 88, 95]

#Crear un grÃ¡fico de lÃ­neas
plt.plot(years, students_eps, color= 'green', marker='o', linestyle = 'solid')

plt.plot(years, students_law, color= 'orange', marker='o', linestyle = 'solid')

#AÃ±adir un tÃ­tulo al grÃ¡fico
plt.title("Estudiantes por aÃ±o")

#AÃ±adir etiqueta al eje x y y (x-axis y y-axis)
plt.ylabel("NÃºmero de Estudiantes")
plt.xlabel("AÃ±o")

#Mostrar
plt.show()
```
